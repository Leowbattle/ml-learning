```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate binomial distribution data using numpy
v = np.random.binomial(n=10, p=0.5, size=10000)
plt.hist(v, bins=30, density=True, alpha=0.5, color='blue')
```

```python
v = (np.random.uniform(low=0, high=1, size=(10000, 10)) > 0.5).sum(axis=1)
plt.hist(v, bins=30, density=True, alpha=0.5, color='blue')
```
