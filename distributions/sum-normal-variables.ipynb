```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m1 = -10
s1 = 1

m2 = 10
s2 = 1

n = 1000000

r1 = np.random.normal(m1, s1, n)
r2 = np.random.normal(m2, s2, n)

s = r1 + r2

plt.hist(r1, bins=30, density=True, alpha=0.5, label=f'N({m1}, {s1})')
plt.hist(r2, bins=30, density=True, alpha=0.5, label=f'N({m2}, {s2})')

plt.hist(s, bins=30, density=True, label=f'Sum')

# s2 = np.random.normal(s.mean(), s.std(), n)
# plt.hist(s2, bins=30, density=True, alpha=0.5, label=f'N({s.mean():.2f}, {s.std():.2f})')

plt.title('Sum of Two Normal Distributions')
plt.xlabel('Value')
plt.ylabel('Density')

plt.legend()
plt.show()
```

```python
nxy = 50
n = 1000

m1 = np.linspace(-10, 10, nxy)
m2 = np.linspace(-10, 10, nxy)

m1, m2 = np.meshgrid(m1, m2)

mean_sum_grid = np.zeros((nxy, nxy))
for i in range(nxy):
    for j in range(nxy):
        r1_ij = np.random.normal(m1[i, j], s1, n)
        r2_ij = np.random.normal(m2[i, j], s2, n)
        mean_sum_grid[j, i] = (r1_ij + r2_ij).mean()

m1_m2_avg = (m1 + m2)

plt.figure(figsize=(8, 6))
plt.imshow(mean_sum_grid - m1_m2_avg, extent=[m1.min(), m1.max(), m2.min(), m2.max()],
           origin='lower', aspect='auto', cmap='viridis')
plt.colorbar(label='Mean of Sum')
plt.xlabel('Mean 1 (m1)')
plt.ylabel('Mean 2 (m2)')
plt.title('Heatmap of Mean(r1 + r2) for Varying Means')
plt.show()
```

```python
ns = 50
n = 100000

m1 = -1
m2 = 1

n_lines = 10
s1 = np.linspace(0, 10, n_lines)
s2 = np.linspace(0, 10, ns)

stds = np.zeros((n_lines, ns))
for j in range(n_lines):
    for i in range(ns):
        r1 = np.random.normal(m1, s1[j], n)
        r2 = np.random.normal(m2, s2[i], n)
        # stds[j, i] = (r1 + r2).std() - np.sqrt(s1[j]**2 + s2[i]**2)
        stds[j, i] = np.sqrt(s1[j]**2 + s2[i]**2)

plt.figure(figsize=(8, 6))
for j in range(n_lines):
    plt.plot(s2, stds[j], label=f'Std 1 = {s1[j]:.2f}')
plt.xlabel('Std 2 (s2)')
plt.ylabel('Std of Sum')
plt.title('Std of Sum vs Std of Second Variable')
plt.legend()
plt.grid()
plt.show()
```

```python
nxy = 50
n = 10000

m1 = -1
m2 = 1

s1 = np.linspace(0, 10, nxy)
s2 = np.linspace(0, 10, nxy)

s1, s2 = np.meshgrid(s1, s2)

std_sum_grid = np.zeros((nxy, nxy))
for i in range(nxy):
    for j in range(nxy):
        r1_ij = np.random.normal(m1, s1[i, j], n)
        r2_ij = np.random.normal(m2, s2[i, j], n)
        std_sum_grid[j, i] = (r1_ij + r2_ij).std()

plt.figure(figsize=(8, 6))
plt.imshow(std_sum_grid, extent=[s1.min(), s1.max(), s2.min(), s2.max()],
           origin='lower', aspect='auto', cmap='viridis')
plt.colorbar(label='Std of Sum')
plt.xlabel('Std 1 (m1)')
plt.ylabel('Std 2 (m2)')
plt.title('Heatmap of Std(r1 + r2) for Varying Means')
plt.show()
```
